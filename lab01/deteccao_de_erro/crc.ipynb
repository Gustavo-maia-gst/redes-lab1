```python
def calcular_crc_manual(dados: str, gerador: str) -> str:
    dados_int = int(dados, 2)
    gerador_int = int(gerador, 2)

    g_len = len(gerador)
    g_degree = len(gerador) - 1
    
    dados_int <<= g_degree
    
    resto = dados_int

    while True:
        r_degree = resto.bit_length() - 1

        if r_degree < g_degree:
            break

        shift = r_degree - g_degree
        resto ^= (gerador_int << shift)

    return bin(resto)[2:].zfill(g_degree)

print(calcular_crc_manual('1101011111', '10011'))
```

```python
config = Configuration(
    width=10,
    polynomial=0b10001001011,  # CRC-3 com gerador x^3 + x + 1
)

calc = Calculator(config)

data = b'eu te amo'
crc = calc.checksum(data)

print(bin(crc)[2:].zfill(3))
```
